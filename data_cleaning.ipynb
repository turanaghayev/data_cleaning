```python
import pandas as pd
import re
import numpy as np
import warnings
warnings.filterwarnings(action = "ignore")
```

```python
df = pd.read_excel("data_cleaning.xlsx", 'Usage', header=None)
```

```python
df
```

```python
df.rename(columns = {0:'column_name'}, inplace = True)
```

```python
df['column_name'] = df['column_name'].str.strip()
```

```python
df['column_name'] = df['column_name'].str.replace('_', '')
```

```python
df
```

```python
df['column_name'].unique()
```

```python
df['Meridian'] = df['column_name'].str.replace(r'^\d*\s*', '')
```

```python
df['Meridian'] = df['Meridian'].str.split(r" ", n = 1, expand = True)[0]
```

```python
df['hour'] = df['column_name'].str.split(r"AM|PM", n = 1, expand = True)[0]
```

```python
df['kwh'] = df['column_name'].str.split(r"2014", n = 1, expand = True)[1]
df['kwh'] = df['kwh'].str.split(r" ", n = 1, expand = True)[0]
df['kwh'] = df['kwh'].str.replace(r'kwh', '')
```

```python
df['a'] = df['column_name'].str.replace(r'(^\d*\s*)', '').str.split(r" ", n = 1, expand = True)[1]
```

```python
df['a'] = df['a'].str.replace('2014', '2014_')
df['date'] = df['a'].str.split(r"_", n = 1, expand = True)[0]
df['date'] = df['date'].str.replace(r'^\s*[A-Z][a-z]+', '')
df['date'] = df['date'].str.replace('th|st|rd|nd', '')
df['date'] = df['date'].str.replace(' ', '')
df['date'] = df['date'].str.replace(r'^[A-Z]+[a-z]*', '')
```

```python
df
```

```python
import datetime
```

```python
df['date'] = pd.to_datetime(df['date'], format="%d-%b-%Y")
```

```python
df
```

```python
df.drop('a', axis=1, inplace= True)
```

```python
df['weekday']=df['date'].dt.day_name()
```

```python
df
```

```python
df.drop(['column_name'], axis=1, inplace= True)
```

```python
df
```

```python
df.info()
```

```python
df['hour']=df['hour'].astype(int)
df['kwh']=df['kwh'].astype(float)
```

```python
df.info()
```

```python
df.sort_values(by=['date','Meridian','hour'], inplace =True)
```

```python
df
```

```python
df = df.reset_index()
df.drop(['index'], axis=1, inplace= True)
```
